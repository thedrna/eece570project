```python
from diffusers import StableDiffusionPipeline
import torch
import os
from PIL import Image
import re
```

# Paths (needs to be changed based on drive location)

```python
prompt_path = "/content/drive/MyDrive/EECE/prompts.txt"
output_dir = "/content/drive/MyDrive/EECE/generated_images_2"
```

```python
with open(prompt_path, "r") as f:
    prompts = [line.strip() for line in f if line.strip()]
```

```python
# prompt: print prompts line by line

for prompt in prompts:
    print(prompt)
```

# Load text-to-image model

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
pipe2 = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    torch_dtype=torch.float16
).to("cuda")
```

# Generating one image for each prompt

```python
for idx, prompt in enumerate(prompts, start=1):
    print(f"Generating image {idx}/{len(prompts)}: {prompt}")
    image = pipe(prompt).images[0]

    filename = f"{idx:02d}_{prompt}.png"
    image_path = os.path.join(output_dir, filename)
    image.save(image_path)
```

```python
for idx, prompt in enumerate(prompts, start=1):
    print(f"Generating image {idx}/{len(prompts)}: {prompt}")
    image = pipe2(prompt).images[0]

    filename = f"{idx:02d}_{prompt}.png"
    image_path = os.path.join(output_dir, filename)
    image.save(image_path)
```
